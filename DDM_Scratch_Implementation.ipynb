# Simulation of Drift Diffusion Model

```python
# Define DDM parameters
dt = 0.01        # Time step (10ms)
T_max = 20.0      # Maximum time duration (5 sec)
c = 0.2          # Noise strength
B = 1
B_1 = 0.8     # Decision boundary

# Define function to compute drift dynamically
def compute_drift(exit_rates, page_value, bounce_rate, traffic_type, visitor_type):
    return weights[0]*exit_rates + weights[1]*page_value + weights[2]*bounce_rate + weights[3]*traffic_type + weights[4]*visitor_type + intercept# Drift function
    #return 0.5 * page_value - 0.5 * bounce_rate  # Drift function

# Drift Diffusion Model simulation
def simulate_ddm(A, c, B, dt, T_max):
    """ Simulates one trial of a Drift Diffusion Model. """
    time_steps = int(T_max / dt)
    x = 0  # Initial evidence
    trajectory = [x]

    for t in range(time_steps):
        dW = np.sqrt(dt) * np.random.randn()  # Gaussian noise
        x += A * dt + c * dW  # Update evidence

        trajectory.append(x)

        # Decision boundary reached
        if x >= B:
            return trajectory, t * dt, 1  # Purchased (1)
        elif x <= -B_1:
            return trajectory, t * dt, 0  # Not purchased (0)

    return trajectory, T_max, None  # No decision
```

```python
# Simulating DDM for multiple users
reaction_times = []
choices = []
num_trials = 5

for _, row in data_1.iloc[:num_trials].iterrows():
    #print(row)
    A = compute_drift(row["ExitRates"], row["PageValues"], row["BounceRates"], row["TrafficType"], row["VisitorType"])  # Compute drift
    traj, rt, choice = simulate_ddm(A, c, B, dt, T_max)  # Run simulation
    reaction_times.append(rt)
    choices.append(choice)
    traj, _, _ = simulate_ddm(A, c, B, dt, T_max)
    plt.plot(np.arange(len(traj)) * dt, traj)
    plt.axhline(B, color='r', linestyle='--', label="Purchase Boundary")
    plt.axhline(-B_1, color='b', linestyle='--', label="No Purchase Boundary")
    plt.xlabel("Time (s)")
    plt.ylabel("Accumulated Evidence")
    plt.title("DDM Simulation for Online Purchase")
    plt.legend()
    plt.show()
```
